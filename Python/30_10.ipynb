```python
num1 = "11"
num2 = "4"
size = len(num1) + len(num2)
product_array = [0] * size
print(product_array)

# if num1 or num2 is zero
if ( num1 == "0" or num2 == "0"):
    ans = "0"

#iterating num1 from right to left
for i in range(len(num1)-1,-1,-1):
    d1 = ord(num1[i]) - ord('0')

    print(f"Digit is at index {i}:{d1}")
     
    for j in range(len(num2)-1,-1,-1):
        d2 = ord(num2[j]) - ord('0')

        print(f"Digit is at index {j}:{d2}")
        
        product = d1 *d2
        index = i + j + 1
        carry_index = i + j

        sum = product + product_array[index]
        product_array[index] = sum % 10
        product_array[carry_index] += sum // 10

print(product_array)

while len(product_array) > 1 and product_array[0] == 0:
    product_array.pop(0)

result = ''.join(map(str,product_array))
print("Product:",result)


```
