1. List and Loop : 
Second Smallest 
Second Largest 
Average of the Even Numbers

```python
Num_input = [10,4,7,9,2,11]
print(Num_input)

n = len(Num_input)
print(f"Length of the Input array = {n}")

largest = second_largest = Num_input[0]
smallest = second_smallest = Num_input[0]

cnt = 0
sum = 0

for value in Num_input:
    
    # Finding smallest from the values in the array 
    if value < smallest:
        second_smallest = smallest
        smallest = value
    
    # If the value is greater than the smallest but smaller than the second smallest then assign that as second smallest 
    elif value > smallest and value < second_smallest and value != smallest:
        second_smallest = value
    
    # Finding Largest from the values in the array 
    if value > largest:
        second_largest = largest
        smallest = value
    
    # If the value is smaller than the largest but greater than the second largest then assign that as second largest
    elif value < largest and value > second_largest and value != largest:
        second_largest = value
    
    if value%2 == 0:
        cnt += 1             # calculating number of even numbers in the array 
        sum += value         # sum of even numbers in the array 

# Typecasting average to the float 
average = float(sum / cnt)

print(f"Second max_element : {second_largest}")
print(f"Second min_element : {second_smallest}")
print(f"average : {average}")
```

2 . Numpy 
    Filling the numpy array with its mean values  

```python
import numpy as np

# Initializing Numpy array 
array = np.array([[1,np.nan,3],[4,5,np.nan],[7,8,9]])

print(array)

# Computing mean for column wise by ignoring nan values and choosing column by axis = 0 
col_mean = np.nanmean(array,axis=0)
print(col_mean)

# Computing indices in the array where the value is nan in the array 
# isnan return the boolean matrix of the array with true false of exxistence of the nan values 
# where return the indices from the boolean input or the expression return as tuple array = [ row_index , column_index]
indices_Nan = np.where(np.isnan(array))
print(indices_Nan)

# Take values from the col_mean array and put in the array where it has nan also by taking indixes as a input 
array[indices_Nan] = np.take(col_mean, indices_Nan[1])

print(array)
```
