3. Minimum window subsequence 

```python
s1 = 'abcdebdde'
s2 = 'bde'

# length of the string s1 and s2
n, m = len(s1), len(s2)

# start = starting index of the best window found
# min_len = stores the length of the best window found so far 
start, min_len = 0, max(n,m)

# iterator for th s2 
j = 0

# Loop over s1 with index i and character ch 
# Enumerators give (i,ch) pair as a output 

for i, ch in enumerate(s1):
    if ch == s2[j]:
        j +=1

        # matched all the character from the s2
        if j == m:
            end = i+1  # marks the end index of the window 
        
        # Shrink the window from the right
            j -= 1     # point last charcter of the s2
            k = i
            while j >= 0:
                if s1[k] == s2[j]:
                    j -= 1
                k -= 1
            k += 1    # k is now start index of minimal window ending at 'end 
            if end - k < min_len:
                start,min_len = k, end-k
            j = 0     # reset j to search for next possible window
 
min_window = s1[start:start + min_len] if min_len != max(n,m) else ""
print(f"Minimum window subsequence :  {min_window}")
```

4. Palindrome pairs 

```python
# input_string = ["bat","tab","cat"]
input_string = ["a","abc","cba"]
result = []

for i in range(len(input_string)):
    for j in range(len(input_string)):
        if i != j:
            combined = input_string[i] + input_string[j]
            if combined == combined[::-1]:
                result.append([i,j])

print(result)
```

5. Letter Combination of a phone number 

```python

# Approach 1 : Iterative Approach 

# iterative Approach to find the letter combination of the phone number 

# initializing the input digits 
digit = "23"  

# Dictionary mapping to the each digit its letter
numbers = {"2":"abc","3":"def","4":"ghi","5":"jkl","6":"mno","7":"pqrs","8":"tuv","9":"wxyz"}

# initializing empty list 
result = [""]

# Base Case : input is empty case 
if not digit:
    print(result)

else:
    for d in digit:
        temp = []
        for combination in result:
            for ch in numbers[d]:
                temp.append(combination + ch)
        result = temp

    print ("Result by the Iterative approach :\n")    
    print (result)

    

# Approach 2 : Recursive Approach 

def letter_combination(digit):

    # Dictionary mapping to the each digit its letter
    numbers = {"2":"abc","3":"def","4":"ghi","5":"jkl","6":"mno","7":"pqrs","8":"tuv","9":"wxyz"}

    # initializing empty list 
    result = []

    # Base Case : input is empty case 
    if not digit:
        print(result)

    # Recursive function (Bracktracking)
    def backtrack(index,current):

        # if current combination is complete 
        if index == len(digit):
            result.append(current)
            return
        
        # Letters for the current digit 
        possible_letters = numbers[digit[index]]

        # All possile letters

        for letter in possible_letters:
            backtrack(index+1,current+letter)
        
    backtrack(0,"")
    print ("\nResult by the Recursive approach :\n")    
    print (result)

letter_combination(digit)
```
