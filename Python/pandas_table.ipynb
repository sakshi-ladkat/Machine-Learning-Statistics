```python
import pandas as pd

person = { "personID" :[1,2,3,4],
    "Lastname":["Ladkat","katare","ghumatkar","Yadav"],
     "Firstname":["Sakshi","Shruti","Shreya","Neha"]}

df_person = pd.DataFrame(person)

print(df_person)

address = { "personID" :[2,3,1,5],
    "Addressid":[1,2,3,4],
     "city":["Mumbai","Pune","Nagpur","Nashik"],
      "State":["MH","MH","MH","MH"]}

df_address = pd.DataFrame(address)
print(df_address)

result = pd.merge(df_person,df_address,on='personID', how="outer")
result = result.drop(columns="personID")

print(result)





log_Numbers =  { "ID" :[1,2,3,4,5,6,7],
    "num":[1,1,1,2,1,2,2],
     }

df_log_Numbers = pd.DataFrame(log_Numbers)
print(df_log_Numbers)

```
