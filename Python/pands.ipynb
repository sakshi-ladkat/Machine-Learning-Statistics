```python
import pandas as pd
print(pd.__version__)

# Reading CSV file using pandas
df = pd.read_csv("bmw_with_missing_data.csv")

# Showing first 5 rows
print(df.head())

# Sum of all missing values (per column)
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Total missing values in the dataset
total_missing = missing_values.sum()
print("Total missing values:", total_missing)

# Remove all rows with missing 'price'
remove_rows = df.dropna(subset=['price'])
print("\nAfter removing rows with missing 'price':")
print(remove_rows)

# Fill missing 'year' values with most common year (mode)
common_year = remove_rows['year'].mode()[0]
print("\nMost common year:", common_year)

remove_rows['year'].fillna(common_year, inplace=True)

print("\nAfter filling missing 'year' values:")
print(remove_rows)

# Average price after cleaning
price_mean = remove_rows['price'].mean()
print("\nAverage price after cleaning:", price_mean)





    




```
